```python
import numpy as np
import pandas as pd
```

## GET THE DATA

```python
column_name = ['user_id','item_id','rating','timestamp']
```

```python
df = pd.read_csv('u.data',sep='\t',names=column_name)
```

```python
df.head()
```

```python
movie_df = pd.read_csv('Movie_Id_Titles')
```

```python
movie_df.head()
```

```python
merged_df = pd.merge(df,movie_df,on='item_id')
```

```python
merged_df.head()
```

```python
ratings = pd.DataFrame(merged_df.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['number_of_rating'] = pd.DataFrame(merged_df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# sns.barplot(x='rating',y='number_of_rating',data=ratings)
sns.distplot(ratings['number_of_rating'],bins=30,kde=False)
```

```python
sns.distplot(ratings['rating'],bins=50,kde=False)
```

```python
sns.jointplot(x='rating',y='number_of_rating',data=ratings,color='red',alpha = 0.3)
```

## Recommending Similar Movies
Now let's create a matrix that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. Note there will be a lot of NaN values, because most people have not seen most of the movies.

```python
movie_matrix = merged_df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
movie_matrix.head(10)
```

```python
ratings.sort_values('number_of_rating',ascending=False).head(10)
```

```python
star_wars_user_rating = movie_matrix['Star Wars (1977)']
```

```python
star_wars_user_rating.head()
```

```python
#Get coorelation values b/w star wars user rating and total matrix user ratings
similar_to_star_wars = movie_matrix.corrwith(star_wars_user_rating)
```

```python
corr_starwars = pd.DataFrame(similar_to_star_wars,columns=['Correlation'])
```

```python
corr_starwars.head()
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars = corr_starwars.join(ratings['number_of_rating'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['number_of_rating'] > 100].sort_values('Correlation',ascending=False).head(10)
```

```python
scream_user_rating = movie_matrix['Scream (1996)']
```

```python
screem_correlation = movie_matrix.corrwith(scream_user_rating)
```

```python
screem_corr_df = pd.DataFrame(screem_correlation,columns=['correlation'])
```

```python
screem_corr_df = screem_corr_df.join(ratings['number_of_rating'])
```

```python
screem_corr_df.head()
```

```python
#Recommended items
screem_corr_df[screem_corr_df['number_of_rating'] > 30].sort_values('correlation',ascending=False).head(5)
```
